```python
import socket

# set up a TCP/IP socket
sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# set up the server address and port
server_address = ('192.168.1.2', 9999)
sock.bind(server_address)

# listen for incoming connections
sock.listen()

while True:
    # wait for a connection
    print('waiting for a connection...')
    connection, client_address = sock.accept()
    print(f'connection from {client_address}')

    # receive the audio data from the client
    audio_data = connection.recv(1024*1024)
    print(f'received {len(audio_data)} bytes of audio data')

    # save the audio data as a file
    with open('audio.wav', 'wb') as f:
        f.write(audio_data)
     
    #load audio file 
    audio_paths2 = ["audio.wav"]
    transcriptions2 = model2.transcribe(audio_paths2)
    print(((transcriptions2[0])['transcription']))
    #########
    # send a response back to the client
    connection.sendall(((transcriptions2[0])['transcription']).encode('utf-8'))

    # close the connection
    connection.close()
```
